```python
import numpy as np 
from matplotlib import pyplot as plt 
import torch
from tqdm.notebook import tqdm
```

```python
x = 0
v = 0
s = 0
t_x = 0.24
t_s = 0.24
lamda = 3
I = 5
x_mon = []
v_mon = []
s_mon = []

dt = 0.01
time = 2000

for i in range(time):
    dx_dt = (-x + v -s + I)/t_x
    x = x + dt * dx_dt
    v = np.tanh(lamda * x)
    ds_dt = (-s + v)/t_s
    s = s + dt * ds_dt

    x_mon.append(x)
    v_mon.append(v)
    s_mon.append(s)
v_avg = np.mean(v_mon)
```

```python
t =np.arange(0,20,0.01)
plt.figure(figsize = (15,3))
plt.plot(t, x_mon, label = 'x')
plt.plot(t, v_mon, label = 'V')
plt.plot(t, s_mon, label = 's')
plt.legend()
```
